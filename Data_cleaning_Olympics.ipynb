```python
def get_names(x):
    # print('here')
    return  x.split('•')[0] +" "+ x.split('•')[1]
bios['Name'] =bios['Used name'].apply(lambda x:get_names(x))

## Another way
# bios['Name'] = bios['Used name].replace('•'," ")
```

```python
import pandas as pd 
def get_measurements(measure): 
    if type(measure) == float and pd.isna(measure): 
        return pd.NA, pd.NA 
    # Return NaN values for height and weight if not entered 
    elif isinstance(measure, str): 
        parts = measure.split('/') 
        if len(parts) >= 2: 
            return parts[0].strip(), parts[1].strip() 
        # Return height and weight 
        else: return pd.NA, pd.NA 
        # Return NaN values if format is invalid 
    else: return pd.NA, pd.NA 
        # Return NaN values for invalid input 
        # Apply the get_measurements function to the 'Measurements' column 
bios[['height', 'weight']] = bios['Measurements'].apply(lambda x: pd.Series(get_measurements(x))) # Drop the 'Measurements' column if needed bios.drop(columns=['Measurements'], inplace=True)
```

```python
## Another way of doing it
bios[['height_cm','weight_kg']] =bios['Measurements'].str.split('/',expand=True)
```

```python

def return_height(x):
    # print('here')
    if type(x) == float and pd.isna(x):  
        return x
    elif 'cm' in x:
        return  x.split(' cm')[0]
    
bios['height'] = bios['height_cm'].apply(lambda x :return_height(x))
bios.head()
```

```python
def return_weight(x):
    # print('here')
    if type(x) == float and pd.isna(x) or x==None:  
        return x
    elif 'kg' in x:

        return  x.split(' kg')[0]
bios['weight'] = bios['weight_kg'].apply(lambda x :return_weight(x))
```
